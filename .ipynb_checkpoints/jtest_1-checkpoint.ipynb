```python
import xarray as xr
import schwabdev as sd
import json
import numpy as np
import pandas as pd
from datetime import datetime

from universes.universe_config import Universe_Config as uc

def create_client():
    """Create a Schwab client with API keys."""
    with open('keys.json', 'r') as f:
        keys = json.load(f)

    return sd.Client(keys['schwab']['app_key'], keys['schwab']['app_secret'])

client = create_client()
```

```python


date_format = "%Y-%m-%d"
start_date = datetime.strptime("2025-01-01",date_format)

hist = client.price_history("F", periodType=None, period=None, frequencyType='minute', frequency=30, startDate=start_date,
                      endDate=None, needExtendedHoursData=True, needPreviousClose=True)

df = pd.DataFrame(hist.json()['candles'])
df['us_date'] = pd.to_datetime(df['datetime'],unit="ms",utc=True).map(lambda x: x.tz_convert("America/new_York"))
df['day_avg_close'] = df.groupby(df['us_date'].dt.date)['close'].transform('mean')
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(df['us_date'],df['close'],color="green")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df['us_date'],df['day_avg_close'],color="green")
plt.show()
```
